```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import json
```

```python
url = 'https://www.tripresso.com'
urlpost = 'https://www.tripresso.com/api_v2/search/ajaxGetResult'
page = 1
GLORIA = '/agency/GLORIA#pageNum='+str(page)
#headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36'}
payload={
    'agency': 'GLORIA'
}
rs = requests.session()
res = rs.post(urlpost,data = payload)
#res2 = rs.get
res.encoding = 'utf-8'
res.text
#soup = BeautifulSoup(res.text,'html.parser')
#soup
```

```python
d = json.loads(res.text)
d.keys()
```

```python
all97 = list(d['groupData'].values())
all97
```

```python
data = pd.DataFrame()
for g in all97:
    for gg in g:
        df = pd.DataFrame([gg.values()])
        data = pd.concat([data,df])
data.columns = gg.keys()
data.reset_index()
gdata = data
gdata
```
