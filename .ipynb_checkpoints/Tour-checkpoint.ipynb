```python
#d['All'][0]
data2 = pd.DataFrame()
n=0
while n < len(d['All']):
    price = d['All'][n]['SaleAm']
    pruduct_name = d['All'][n]['GrupSnm']
    EstmTotqt = d['All'][n]['EstmTotqt']
    SaleYqt = d['All'][n]['SaleYqt']
    pruduct_id = d['All'][n]['GrupCd']
    date = d['All'][n]['OrderDl']
    days = d['All'][n]['GrupLn']
    df = pd.DataFrame([[pruduct_id,date,days,pruduct_name,SaleYqt,EstmTotqt,price]])
    data2 = pd.concat([data2,df])
    n += 1
data2.columns = [["pruduct_id","date","days","pruduct_name","SaleYqt","EstmTotqt","price"]]
data2
```
